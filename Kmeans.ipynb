```python
import numpy as np
from random import shuffle
    
class Cluster:
    centroid = []
    samples = []
    def __init__(self, model, numSamples):
        self.centroid = []
        self.samples = []
        for b in model.bounds:
            self.centroid.append( np.random.uniform(b[0], b[1], 1)[0] )
        print self.centroid
        sigma = 0.06
        self.samples = np.random.multivariate_normal(self.centroid, sigma * np.eye(len(self.centroid)), numSamples)

class Model:
    clusters = []
    bounds = [ [-1, 1], [-1, 1] ]
    def __init__(self, numClusters, numSamples):
        self.clusters = []
        for n in range(numClusters):
            newCluster = Cluster(self, numSamples)
            self.clusters.append( newCluster )
            
    def getSamples(self):
        samples = []
        for c in self.clusters:
            for s in c.samples:
                samples.append(s)
                
        # shuffle the samples
        shuffle(samples)

        return samples
    
    def plot(self):
        markers = ['ro', 'bs', 'm^', 'y*', 'b<', 'y>']
        for cn, c in enumerate(self.clusters):
            for p in c.samples:
                plot(p[0], p[1], markers[cn])
                grid(True)
        title('True Model')

class KMeanSolver:
    bounds = []
    k = 1
    samples = []
    centroids = []
    ndim = 0
    iteration = 0
    
    def distance(self, sample, centroid):
        v = np.array(sample) - np.array(centroid)
        d = 0;
        for n,p in enumerate(v):
            d = d + p**2
        return d
    
    def assignSamplesToClusters(self):
        # Make the cluster list have correct dimensions
        # so that we can append
        self.clusters = []
        for n in range(self.k):
            self.clusters.append([])
        
        for s in self.samples:
            dlist = []
            for c in self.centroids:
                dlist.append(self.distance(s, c))
            #print dlist
            val, idx = min((val, idx) for (idx, val) in enumerate(dlist))
            #print "best cluster = " + str(idx)
            self.clusters[idx].append(s)
    
    def updateCentroids(self):
        for idx, c in enumerate(self.clusters):
            newCentroid = []
            for n in range(self.ndim):
                cen = 0.0
                if len(c) == 0:
                    break;
                for s in c:
                    cen = cen + s[n]
                cen = cen / len(c)
                newCentroid.append(cen)
            self.centroids[idx] = newCentroid
    
    def iterate(self):
        # Calculate and update the centroids
        self.updateCentroids()
        
        # Assign each sample to a cluster
        self.assignSamplesToClusters()
            
        
        
        self.iteration = self.iteration + 1
    
    def plot(self):
        markers = ['ro', 'bs', 'm^', 'y*', 'b<', 'y>']
        for cn, c in enumerate(self.clusters):
            if len(c) == 0:
                continue
            
            for p in c:
                plot(p[0], p[1], markers[cn])
                grid(True)
            plot(self.centroids[cn][0], self.centroids[cn][1], 'mx')
        title('K-Mean Solver: iteration = ' + str(self.iteration))
        
    def __init__(self, k, samples):
        self.bounds = []
        self.k = k
        self.samples = samples
        self.centroids = []
        self.clusters = []
        self.iteration = 0
        
        # Determine the bounds
        self.ndim = len(samples[0])
        bounds = list()
        for n in range(self.ndim):
            x = [s[n] for s in samples]
            self.bounds.append([min(x), max(x)])
            
        # Pick k of the samples to be the initial centroids
        initialIndex = np.random.randint(0, len(self.samples), self.k)
        for n in initialIndex:
            self.centroids.append( self.samples[n] )
        
        # Make the cluster list have correct dimensions
        # so that we can append
        self.clusters = []
        for n in range(self.k):
            self.clusters.append([])
        
        # Assign each sample to a cluster
        self.assignSamplesToClusters()
            
        # Calculate and update the centroids
        #self.updateCentroids()
        
```
